```python
import pandas as pd
```

**Change csv file directory**

```python
dataset = pd.read_csv('feeds.csv')
dataset['created_at'] = pd.to_datetime(dataset['created_at'], format='ISO8601')
dataset.drop(['latitude','longitude','elevation','status','field3'],axis=1,inplace=True)
dataset.rename(columns = {'field1':'temp','field2':'ph'}, inplace = True)

# # Get the ammonia from obtained temp and pH values by using the formula from https://ieeexplore-ieee-org.dlsu.idm.oclc.org/stamp/stamp.jsp?tp=&arnumber=8269494
dataset['ammonia'] = (10**(0.0901821+(2729.92/(273.15-dataset['temp']))-dataset['ph']))**-1

# # Filters unwanted values
filtered_df = dataset[(dataset['temp']>25) & (dataset['temp']<40) & (dataset['ph']>5.5) & (dataset['ph']<9.5)].reset_index() # Filter temp values between 25 and 40 and pH values between 5.5 and 9.5
filtered_df.index = filtered_df.index.set_names(['new_index'])
```

```python
filtered_df.to_csv('new feeds testing.csv')
```
